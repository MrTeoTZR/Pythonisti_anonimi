Per accedere alle info di Wikipedia direttamente dal notebook, si usa la libreria "wikipedia".
Nel dettaglio, useremo principalmente i comandi "search" e "suggestion".
"search"= restituisce il risultato di ricerca di una query.

La ricerca fatta con 'search' possiamo filtrarla usando parametri come 'results' (scrivendo il numero massimo di risultati) e 'suggestion' (se 'True', restituisce una tupla formata da risultato + un suggerimento).

```python
import wikipedia as wiki
print (wiki.search("Worldwide Crop Production", results = 10, suggestion = True))
```

Il comando 'suggest' è fatto per facilitare le richieste (ma non funziona un granchè): se individua un titolo di una pagina sul web leggendo la query scritta male, suggerisce la parola di ricerca corretta. Altrimenti, restituisce 'none'.

```python
print (wiki.suggest ("Worlwide crop productn"))
```

Il comando 'summary' fornisce una spiegazione della parola cercata sul web.

```python
print (wiki.summary ("Worldwide Crop Production"))
```

Il comando 'set_lang' ("nome_lingua") permette di tradurre il testo del risultato della query nella lingua che vogliamo. E' giusto una curiosità però: meglio lasciare la lingua settata su "en", altrimenti non funziona più nulla...

```python
wiki.set_lang("fr")
print (wiki.search ("Worldwide Croop Production"))
```

L'API ci permette di accedere anche direttamente alla pagina di Wikipedia. Il comando da usare è 'page' e ci permette di accedere al titolo, all'URL, al contenuto, ad immagini e a link completi della pagina.

```python
agricoltura = wiki.page ("Crop Production")
print (agricoltura.title)
```

```python
print (agricoltura.url)
```

```python
print (agricoltura.content)
```

```python
print (agricoltura.images)
```

```python
print (agricoltura.links)
```
